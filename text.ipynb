```python
import pandas as pd
```

```python
dr_Dataset = pd.read_excel("dr.xlsx")
print(dr_Dataset)
```

```python
df = pd.DataFrame({
    "Kitap Adı": dr_Dataset["Kitap Adı"],
    "Yazar":  dr_Dataset["Yazar"],
    "Yayinevi": dr_Dataset["Yayinevi"],
    "Fiyat": dr_Dataset["Fiyat"],
}, columns=["Kitap Adı","Yazar","Yayinevi","Fiyat"])
df
```

```python
KitapAdiMaxUzunluk = max(df["Kitap Adı"].apply(len))
YazarMaxUzunluk = max(df["Yazar"].apply(len))
YayineviMaxUzunluk = max(df["Yayinevi"].apply(len))
FiyatMaxUzunluk = max(df["Fiyat"].apply(len))
```

```python
writer = pd.ExcelWriter("TEST_dr2.xlsx", engine="openpyxl")
df.to_excel(writer, index=False, encoding="utf-8-sig")
sheet = writer.sheets["Sheet1"]
print(sheet)
sheet.column_dimensions["A"].width = KitapAdiMaxUzunluk
sheet.column_dimensions["B"].width = YazarMaxUzunluk
sheet.column_dimensions["C"].width = YayineviMaxUzunluk
sheet.column_dimensions["D"].width = FiyatMaxUzunluk
writer.save()
print(sheet)
```

```python
hierarchy = list(zip(df["Yayinevi"],df["Yazar"],df["Fiyat"]))
hierarchy = pd.MultiIndex.from_tuples(hierarchy)
hierarchy
```

```python
df.set_index(hierarchy, inplace=True)
df.index.names = ["Yayinevleri","Yazarlar","Fiyatlar"]
df
```

```python
df.drop(["Yayinevi","Yazar","Fiyat"], axis=1, inplace=True)
df = df.sort_index()
df
```

```python
df.to_excel("TEST_dr.xlsx", index=False, encoding="utf-8-sig", freeze_panes=(1,0))
```
